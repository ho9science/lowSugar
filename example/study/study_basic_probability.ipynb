## 조건부 확률

### 아들 딸 문제
1. 각 아이가 딸이거나 아들일 확률은 동일
2. 둘째의 성별은 첫째의 성별과 독립

가능한 경우의 수 : (딸, 딸), (딸, 아들), (아들, 딸), (아들, 아들)

첫째가 딸일 경우(사건A) 두 아이가 모두 딸일(사건B) 확률
P(B|G) = P(B,G)/P(G) = P(B)/P(G) = 1/2

딸이 최소 한명인 경우(사건L), 두 아이가 모두 딸일 확률
P(B|L) = P(B,L)/P(L) = P(B)/P(L) = 1/3

```python
import random

def random_kid():
    return random.choice(["boy", "girl"])

both_girls = 0
older_girl = 0
either_girl = 0
random.seed(0)

for _ in range(10000):
    younger = random_kid()
    older = random_kid()
    if older == "girl":
        older_girl +=1
    if older == "girl" and younger == "girl":
        both_girls += 1
    if older == "girl" or younger == "girl":
        either_girl += 1

both_girls/older_girl
```

```python
both_girls / either_girl
```
